### 读取数据

```python
file = 'D:/Meteorology/雷达基数据/Z_RADR_I_Z9576_20180629043900_O_DOR_SA_CAP.bin.bz2'
f = cinrad.io.CinradReader(file)
```

### 衍生产品计算

#### 获取整个体扫的反射率数据

```python
a = [f.reflectivity(i, 230) for i in f.anglelist_r]
```

#### 使用easycalc模块计算

```python
et = cinrad.easycalc.quick_et(a)
```
